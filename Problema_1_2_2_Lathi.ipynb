#Problema 1.2-2 Lathi

## Planteamiento del problema:

Se resuelve el problema 1.2-2 del libro lathi.

1.2-2 For the signal x(t) illustrated in Fig. P1.2-2,sketch.

```python
import numpy as np
import matplotlib.pyplot as plt

print("Problema 1.2-2 Lathi\n");
print("Se resuelve el problema 1.2-2 del libro lathi.\n");
print("1.2-2 For the signal x(t) illustrated in Fig. P1.2-2,sketch.\n");
t = np.linspace(-5, 4)
y = np.piecewise(t, [t<-4,(t>=-4) & (t<0),(t>=0) & (t<2),t>=2], [lambda t: 0, lambda t:-t, lambda t: t,lambda t: 0])
plt.title('Figure P1.2-1', fontsize = 15)
plt.xlabel("t→", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:blue'})
plt.xlim(-5, 4)
plt.plot(t, y,color="r")
plt.grid()
plt.show()
print("\n");
print("\t(a) x(t-4).\n");
print("\t(b) x(t/1.5).\n");
print("\t(c) x(-t).\n");
print("\t(d) x(2t-4).\n");
print("\t(e) x(2-t).\n");
print("\n");
```

# Inciso (a).

```python
print("\n");
plt.title('(a) x(t-4).', fontsize = 15)
plt.xlabel("t→", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:blue'})
plt.xlim(-9, 0)
plt.plot(t-4, y,color="y")
plt.grid()
plt.show()
```

# Inciso (b).

```python
print("\n");
plt.title('(b) x(t/1.5).', fontsize = 15)
plt.xlabel("t→", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:blue'})
plt.xlim(-3, 2)
plt.plot(t/1.5, y,color="b")
plt.grid()
plt.show()
```

# Inciso (c).

```python
print("\n");
plt.title('(c) x(-t).', fontsize = 15)
plt.xlabel("t→", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:blue'})
plt.xlim(-4, 5)
plt.plot(-t, y,color="pink")
plt.grid()
plt.show()
```

# Inciso (d)

```python
print("\n");
plt.title('(d) x(2t-4).', fontsize = 15)
plt.xlabel("t→", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:blue'})
plt.xlim(-13.5, 2.5)
plt.plot(2*t-4, y,color="orange")
plt.grid()
plt.show()
```

# Inciso (e)

```python
print("\n");
plt.title('(e) x(2-t).', fontsize = 15)
plt.xlabel("t→", fontdict = {'fontsize':12, 'fontweight':'bold', 'color':'tab:blue'})
plt.xlim(-2, 7)
plt.plot(2-t, y)
plt.grid()
plt.show()
```
